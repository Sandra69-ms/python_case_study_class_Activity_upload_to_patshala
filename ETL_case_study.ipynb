4. Convert time to datetime

```python
data['time']=pd.to_datetime(data['time'])
```

```python
data.info()
```

5. Drop records with missing latitude, longitude, mag

```python
data.isnull().sum()
```

```python

data = data.dropna(subset=['latitude', 'longitude', 'mag'])

print(f"Original records: {len(data)}")
print(f"After cleaning: {len(data)}")
```

```python
data.info()
```

6. Filter only earthquakes with magnitude >= 4.0

```python

 data[data['mag'] >= 4.0]

```

```python
print(data.head())
```

7.Add a column day_of_week from time

```python
data['day_of_week'] =data['time'].dt.day_name()
```

```python
print(data.columns)
```

8. Create a column severity_level based on magnitude:

â—‹ <4.0: "Low"

â—‹ 4.0 - 6.0: "Moderate"

â—‹ 6.0+: "High"


```python
conditions = [
    data['mag'] < 4.0,
    (data['mag'] >= 4.0) & (data['mag'] < 6.0),
    data['mag'] >= 6.0
]

choices = ['Low', 'Moderate', 'High'];


data['severity_level'] = np.select(conditions, choices, default='Unknown')

print(data[['mag', 'severity_level']].head())
```

9. Count number of earthquakes per place

```python
place_counts = data['place'].value_counts()
print(place_counts)
```

10.  Compute average magnitude and max depth per day

```python
earthquake_summary = data.groupby(data['time'].dt.date).agg({'depth':'max', 'mag':'mean'}).reset_index()
earthquake_summary.columns = ['date', 'max_depth', 'average_magnitude']
```

13 .  A short explanation of your process

The process involved the following steps:

1. **Extraction**: Data was loaded from a CSV file hosted on the USGS website. Basic information about the dataset, such as the number of records and unique locations, was identified.
2. **Transformation**: The 'time' column was converted to datetime objects. Records with missing values in essential columns (latitude, longitude, and magnitude) were dropped. The data was then filtered to include only earthquakes with a magnitude of 4.0 or greater. New columns for 'day_of_week' and 'severity_level' were added based on the 'time' and 'mag' columns, respectively. The number of earthquakes per place was counted, and the daily maximum depth and average magnitude were computed.
3. **Loading**: The cleaned and summarized data were saved to CSV files. Finally, the cleaned data was stored in a SQLite database.
