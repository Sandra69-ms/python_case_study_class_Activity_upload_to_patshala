17,Create a column that categorizes movies by rating:

a) Rating < 6: Poor

b) 6 ≤ Rating < 7.5: Average

c) Rating ≥ 7.5: Excellent

```python
conditions=[df['Rating'] < 6 ,(df['Rating']>=6) & (df['Rating']<=7.5) ,df['Rating'] >=7.5 ]
values =["poor",'average','excellent']
df['movie_rating']= np.select(conditions,values,default='UNKNOWN')
df.head(10)
```

18,Create a custom function to score movies based on:

i) Apply row['Rating'] * 10 + row['Revenue (Millions)'] as High score

Apply a function to dataset to assign the score

```python
def calculate_high_score(row):
    return row['Rating'] * 10 + row['Revenue (Millions)']

df['High_Score'] = df.apply(calculate_high_score, axis=1)

print(df[['Title', 'Rating', 'Revenue (Millions)', 'High_Score']].head())
```
