Identify missing values in each column.

```python
# identify the missing value from columns
house_data.isna().sum()
```

```python
# removal of  null values from the target  here sale price is target

house_data = house_data.dropna(subset=['Sale Price'])
```

```python
house_data.head()
```

```python
from scipy.stats import skew
```

```python
skewness = house_data["No of Bathrooms"].skew
skewness()
```

```python
# Fill missing values in 'No of Bathrooms' with the median
house_data.loc[:,"No of Bathrooms"]=house_data['No of Bathrooms'].fillna(house_data['No of Bathrooms'].median()) #filling with median
```

```python
house_data.loc[:,"Flat Area (in Sqft)"]=house_data['Flat Area (in Sqft)'].fillna((house_data['Area of the House from Basement (in Sqft)'])+(house_data['Basement Area (in Sqft)']))
# here flat area is filled using the sum of area of house from basement and basement area
```

```python
house_data.loc[:,"Lot Area (in Sqft)"] = house_data['Lot Area (in Sqft)'].fillna(house_data["Lot Area after Renovation (in Sqft)"])

 # here lot area is filled using lot area after renovation  from house data
```

```python
house_data.isna().sum()
```

```python
house_data = house_data.drop("No of Times Visited", axis =1)
```

```python
house_data =house_data.dropna(subset= ["Zipcode"])
```

```python
house_data = house_data.drop(["Latitude","Longitude"],axis =1)
```

```python

house_data = house_data.dropna(subset=['Living Area after Renovation (in Sqft)'])
```

```python
house_data.loc[:,"Area of the House from Basement (in Sqft)"]=house_data['Area of the House from Basement (in Sqft)'].fillna((house_data['Flat Area (in Sqft)'])-(house_data['Basement Area (in Sqft)']))

#here the basement area from house is filled using flat area is subracted from the basement area
```

```python
house_data = house_data.drop(["Basement Area (in Sqft)","Area of the House from Basement (in Sqft)"],axis=1)
```

```python
house_data = house_data.drop(["Renovated Year","Lot Area (in Sqft)"],axis=1)
```

```python
# iQR method to find outlier for no of bedroom

```

```python
q1 = np.quantile(house_data['No of Bedrooms'],0.25)
q3 = np.quantile(house_data['No of Bedrooms'],0.75)
iqr = q3 - q1
upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)
print(upper_bound)
print(lower_bound)
print(iqr)
```

```python
bedroom_outliers = []
for i in house_data['No of Bedrooms']:
  if i > upper_bound or i < lower_bound:
    bedroom_outliers.append(i)
```

```python
house_data['No of Bedrooms'] = house_data['No of Bedrooms'].clip(lower = lower_bound,upper= upper_bound)
```

```python
plt.boxplot(house_data['No of Bedrooms'])
```

```python
#handling outlier in the Living Area after Renovation
skewness=house_data['Living Area after Renovation (in Sqft)'].skew
skewness()
```

```python
from scipy.stats.mstats import winsorize

# Calculate the winsorized values (capping at 5% from both ends)
house_data['Living Area after Renovation (in Sqft)'] = winsorize(house_data['Living Area after Renovation (in Sqft)'], limits=[0.05, 0.05])

# Check the skewness after winsorizing
skewness = house_data['Living Area after Renovation (in Sqft)'].skew()
print(f"Skewness after winsorizing: {skewness}")

# Plot a boxplot to visualize the effect of winsorizing
plt.boxplot(house_data['Living Area after Renovation (in Sqft)'])
plt.title('Living Area after Renovation (in Sqft) after Winsorizing')
plt.show()
```

```python
#handling outlier in Lot Area after Renovation (in Sqft)

skewness=house_data['Lot Area after Renovation (in Sqft)'].skew
skewness()
```

```python
house_data['Lot Area after Renovation (in Sqft)']=np.log(house_data['Lot Area after Renovation (in Sqft)'])
```
