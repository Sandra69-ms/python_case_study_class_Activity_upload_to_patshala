# univarite

categorical data

```python
# The distribution of profit

plt.figure(figsize=(8,10))
sns.histplot(data["Sub-Category"],bins=40, color="green")
plt.title("Distribution of sub category")
plt.xlabel("sub category")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()


#Here "Sub-Category" is a categorical variable, and a count plot shows how often each category occurs.

#The chart reveals which sub-categories dominate sales
 #(e.g., Binders and Paper have the highest frequency, while Copiers are the least frequent).
```

Numerical data

```python

sales_by_category = data.groupby("Category")["Sales"].sum()

plt.figure(figsize=(6,6))
plt.pie(
    sales_by_category,
    labels=sales_by_category.index,
    autopct="%.1f%%",
    startangle=90
)
plt.title("Sales Distribution by Category")
plt.show()


# Each slice shows the proportion of total Sales contributed by that Category.

# here Technology accounts for ~35% of total sales, followed by Office Supplies at 34% and Furniture at 31%.
```

# Bivarite

numeric vs numeric

```python
# sales vs profit

plt.figure(figsize=(7,5))
sns.scatterplot(x="Sales", y="Profit", data=data, hue="Category", alpha=0.7)
plt.title("Bivariate Analysis: Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.xticks(rotation= 45)
plt.yticks(rotation= 45)
plt.show()


# Here shows Sales vs Profit because both are numerical variables, and scatterplots reveal whether higher sales lead to higher profits.

# The plot shows that while most sales generate profit,
#some large transactions result in losses ,especially in certain categories like Furniture.
```

```python


# Ensure 'Order Date' is datetime and set as index
data['Order Date'] = pd.to_datetime(data['Order Date'], errors='coerce')
data.set_index('Order Date', inplace=True)

# Resample to monthly sums
monthly_data = data[['Sales', 'Profit']].resample('M').sum()

# Plot both Sales and Profit on the same line plot
plt.figure(figsize=(10,5))
plt.plot(monthly_data.index, monthly_data['Sales'], marker='o', label='Sales', color='blue')
plt.plot(monthly_data.index, monthly_data['Profit'], marker='o', label='Profit', color='green')
plt.title('Monthly Sales vs Profit Trend')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
plt.show()


# The trend of Sales and Profit over time.

#How closely Profit follows Sales, highlighting seasonal patterns or months where high sales did not generate high profit.
```

categorical vs numerical

```python
 #profit by category


 plt.figure(figsize=(6,4))
sns.boxplot(x="Category", y="Profit", data=data)
plt.title("Bivariate Analysis: Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.xticks(rotation= 45)
plt.yticks(rotation = 45)
plt.show()


# Higher sales generally lead to higher profit, showing a positive trend overall.

#Some categories (like Furniture) have large sales with low or negative profit, indicating higher costs or discounts.
```

category vs category

```python
# region vs segment

plt.figure(figsize=(8,5))
sns.countplot(x="Region", hue="Segment", data=data)
plt.title("Customer Segment by Region")
plt.xlabel("Region")
plt.ylabel("Segment")
plt.xticks(rotation =45)
plt.show()

#The plot shows how customer segments are distributed across different regions.
# It reveals which segments dominate in each region and highlights regional differences in customer composition.

```

```python
# shipmode vs segment frequencies

contingency_table = pd.crosstab(data['Ship Mode'], data['Segment'])
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Ship Mode vs. Segment Frequencies')
plt.show()


# The heatmap shows the frequency of each combination of Ship Mode and Segment,
# revealing which shipping methods are most popular for each customer segment.
# It highlights that certain segments prefer specific ship modes, indicating patterns in delivery preferences.
```

# Multivarite

```python
#Sales vs Profit by Category and Discount

sns.scatterplot(data=data, x='Sales', y='Profit', hue='Category', style='Segment')
plt.title('Sales vs. Profit by Category and Segment')
plt.show()


#The plot shows that higher sales generally correspond to higher profits.
#It also highlights differences across categories and how varying discounts affect profitability.
```
